```python
import numpy as np
import pandas as pd

df = pd.read_csv("dataset2024.csv")
df = df.fillna("")
df.head()
```

```python
def preprocessing(corpus):
    return corpus.split()
import contractions
df["text"] = df["text"].astype(str).apply(lambda x: contractions.fix(x))
df["pre"] = df["pre"].astype(str).apply(lambda x: contractions.fix(x))
df["post"] = df["post"].astype(str).apply(lambda x: contractions.fix(x))
```

```python
df["text_clean"] = df["text"].astype(str).str.lower()
df["text_clean"].head()
```

```python
df["pre"] = df["pre"].astype(str).str.lower()
df["post"] = df["post"].astype(str).str.lower()
df.head()
```

```python
import re
def remove_punct(s: str) -> str:
    if pd.isna(s):
        return ""
    s = str(s).lower()
    s = re.sub(r'(\d+)\.(\d+)', r'\1<dot>\2', s)
    s = re.sub(r'(\d+)-(\d+)', r'\1 to \2', s)
    s = re.sub(r'[^a-z0-9\s<dot>]', '', s)
    s = s.replace("<dot>", ".")
    s = re.sub(r'\s+', ' ', s).strip()
    return s


df["text_clean"] = df["text_clean"].apply(remove_punct)
df["pre"] = df["pre"].apply(remove_punct)
df["post"] = df["post"].apply(remove_punct)

df.head()
```

```python
def indexing_word(words):
    word_to_idx = {word:idx for idx,word in enumerate(words)}
    idx_to_word = {idx:word for idx,word in enumerate(words)}
    return word_to_idx, idx_to_word
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

def remove_stopwords_spacy(text):
    doc = nlp(text)
    return [token.text for token in doc if not token.is_stop]
```

```python
def lemmatize_spacy(text):
    doc = nlp(text)
    return " ".join([token.lemma_ for token in doc])
```

```python
df["text_lemma"] = df["text_clean"].astype(str).apply(lemmatize_spacy)
df["pre"] = df["pre"].astype(str).apply(lemmatize_spacy)
df["post"] = df["post"].astype(str).apply(lemmatize_spacy)

df.head()
```
