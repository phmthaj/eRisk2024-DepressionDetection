```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

df_pred = pd.read_csv("test_predictions.csv")

y_true = df_pred["relevant"]
y_pred = df_pred["predicted"]

acc = accuracy_score(y_true, y_pred)

p, r, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="binary")

print(f"Accuracy : {acc:.4f}")
print(f"Precision: {p:.4f}")
print(f"Recall   : {r:.4f}")
print(f"F1-score : {f1:.4f}")

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
disp.plot(cmap="Blues")
plt.title("Confusion Matrix")
plt.show()
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

df = pd.read_csv("test_predictions_2.csv")

results = []

for qid, group in df.groupby("query"):
    y_true = group["relevant"]
    y_pred = group["predicted"]

    acc = accuracy_score(y_true, y_pred)
    p, r, f1, _ = precision_recall_fscore_support(
        y_true, y_pred, average="binary", zero_division=0
    )

    results.append({
        "query": qid,
        "accuracy": acc,
        "precision": p,
        "recall": r,
        "f1": f1,
        "support": len(group)
    })

results_df = pd.DataFrame(results)
print(results_df)

```
